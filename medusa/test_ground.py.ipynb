```python
import requests
import pprint
import os
from github import Github

GITHUB_USERNAME = os.environ.get("GITHUB_USERNAME")
GITHUB_PASSWORD = os.environ.get("GITHUB_PASSWORD")

session = Github(GITHUB_USERNAME, GITHUB_PASSWORD)
```

```python
repos = session.search_repositories("utils+language:python")
```

```python
repos.totalCount

for page in (repos.get_page(num) for num in range(repos.totalCount)):
     pprint.pprint(page)
```

```python
code_files = []
for repo in repos["items"]:
    full_name = repo["full_name"]
    new_url = f"https://api.github.com/search/code?q=def+language:python+in:file+repository={full_name}"
    print(new_url)
    code_files.append(
        requests.get(new_url).json())
```
